# Preprocessing von Text Daten

```python
# Document = in sich geschlossener Text, den wir analysieren wollen
```

```python
from textblob import TextBlob
from string import punctuation, ascii_uppercase, ascii_lowercase, ascii_letters
from nltk.stem import PorterStemmer
```

```python
# !python -m textblob.download_corpora
```

### Daten laden

```python
with open("output/quotes_10.txt", "r", encoding="utf-8") as file:
    content = file.read()
```

```python
content
```

### Vorbereiten

```python
# alles in Kleinbuchstaben umwandeln
content = content.lower()
```

```python
new_content = ""
for char in content:
    if char not in punctuation + "“" + "”":
        new_content += char # += ist das selbe wie new_content = new_content + char

content = new_content # jetzt haben wir einen reinen ascii text ohne sonderzeichen
```

```python
content
```

### Text Blob

```python
blob = TextBlob(content)
blob.words
```

```python
lemmas = [word.lemmatize() for word in blob.words]
# lemmatize -> Was ist der ursprüngliche Wortstamm, Mehrzahl ->  Singular, z.B. feet -> foot
# stemming -> man schneidet das Wort einfach ab, z.B. running -> run
```

```python
stemmer = PorterStemmer()
stemmer.stem("feet")
stems = [stemmer.stem(word) for word in blob.words]
```

```python
print(lemmas)
print(stems)
```
