# Quotes to Scrape

## Requesting Web-Content via HTTP

```python
from requests import get
from bs4 import BeautifulSoup
from time import sleep # pausiert unseren Code für eine bestimmte Zeit (damit der Server nicht überlastet wird)

USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
```

```python
get(base_url)
# erhalten ein Response-Objekt mit Statuscode 200 (OK); 
# Es gibt auch andere Statuscodes, z.B. 404 (Not Found)
```

## Parsing Web-Content in Python

```python
base_url = "https://quotes.toscrape.com/page/"

corpus = []

for page in range(1, 10):
    url = base_url + str(page)
    
    print("Requesting", url)

    response = get(url, headers={"User-Agent": USER_AGENT})


    soup = BeautifulSoup(
        response.content,
        "html.parser"
    )


    quotes = soup.find_all(
        "div", 
        attrs={"class": "quote"}
    )
    # Find_all findet alle Attribute
    # find_next findet nur das erste Attribut


    for quote in quotes:
        
        textquote = quote.find_next(
            "span", 
            attrs={"class": "text"}
        )
        
        corpus.append(textquote.text)


    sleep(5)  # pausiert den Code für 5 Sekunden
```

```python
for index, quote in enumerate(corpus):
    
    with open(f"output/quotes_{index}.txt", "w", encoding="utf-8") as file:
        file.write(quote)
```
